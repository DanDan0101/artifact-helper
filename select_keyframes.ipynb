```python
from pathlib import Path

import numpy as np
from matplotlib import pyplot as plt
from tqdm.notebook import tqdm

import imageio.v3 as iio
```

```python
ARTIFACT_LOC = (202, 2208)
ARTIFACT_SIZE = (1754, 830)
```

```python
dir = "samples"
file_name = "artifacts_short.mkv"
file_path = Path(f"./{dir}/{file_name}").resolve()
```

```python
def n_frames(file_path):
    metadata = iio.immeta(file_path, plugin="pyav")
    fps = metadata["fps"]
    duration = metadata["DURATION"]
    hours, minutes, seconds = map(float, duration.split(":"))
    seconds += hours * 3600 + minutes * 60
    # Idk why 2 frames are missing, fudge factor results from testing a single video
    # TODO: test more videos to see if the -2 is consistent
    return int(np.rint(seconds * fps) - 2)
```

```python
def crop_frame(frame, loc=ARTIFACT_LOC, size=ARTIFACT_SIZE):
    return frame[loc[0] : loc[0] + size[0], loc[1] : loc[1] + size[1], :]
```

```python
for frame in tqdm(
    iio.imiter(file_path, plugin="pyav", thread_type="FRAME"), total=n_frames(file_path)
):
    cropped_frame = crop_frame(frame)
```
